```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

#ambil data
df = pd.read_csv('data_1.csv')
df
```

```python
#cek data type
df.info()
```

```python
#cek banyaknya data
df.shape
```

```python
#cek jumlah data kosong
df.isnull().sum()
```

```python
#ganti data dengan ukuran pusat data
df['Umur'].fillna(df['Umur'].mean(), inplace=True)
df['Pendapatan'].fillna(df['Pendapatan'].median(), inplace=True)
df
```

```python
#label encode
le = LabelEncoder()
df['Tingkat Pendidikan'] = le.fit_transform(df['Tingkat Pendidikan'])
df
```

```python
#One-Hot encode
df = pd.get_dummies(df, columns=['Jenis Kelamin','Status Perkawinan'])
df
```

```python
#transofrmasi data tipe numerik
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['Pendapatan_Scaled'] = scaler.fit_transform(df[['Pendapatan']])
df
```
